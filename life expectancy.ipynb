```python
# renaming the columns
dataset.columns = ['country', 'year', 'status', 'life_expectancy', 'adult_mortality', 'infant_deaths', 'alcohol',
                   'percentage_expenditure', 'hepatitis_B', 'measles', 'bmi', 'under_5_deaths', 'polio', 'total_expenditure',
                   'diphtheria', 'hiv_aids', 'gdp', 'population', 'thin_1-19_years','thin_5-9_years',
                   'income_comp_of_resources', 'schooling']
```

```python
# checking the number of unique items in the dataset
dataset.apply(lambda x: len(x.unique()))
```

```python
# checking for duplicate rows
dataset.duplicated().sum()
```

```python
# checking for outliers
datasetn = dataset.drop(columns=["year","country","status"])
def outlier_count(col, data = datasetn):
    
    print("\n"+15*'-' + col + 15*'-'+"\n")
    
    q75, q25 = np.percentile(data[col], [75, 25])
    iqr = q75 - q25
    min_val = q25 - (iqr*1.5)
    max_val = q75 + (iqr*1.5)
    outlier_count = len(np.where((data[col] > max_val) | (data[col] < min_val))[0])
    outlier_percent = round(outlier_count/len(data[col])*100, 2)
    print('Number of outliers: {}'.format(outlier_count))
    print('Percent of data that is outlier: {}%'.format(outlier_percent))

for col in datasetn:
    outlier_count(col)
```

```python
# converting categorical columns to numeric
from sklearn.preprocessing import LabelEncoder
le= LabelEncoder()
dataset['status']=le.fit_transform(dataset['status'])
dataset['country']= le.fit_transform(dataset['country'])
```

```python
dataset
```

```python
from sklearn.preprocessing import RobustScaler
scaler = RobustScaler()
scaler.fit(dataset.drop('life_expectancy', axis=1))
scaled_features = scaler.transform(dataset.drop('life_expectancy', axis=1))
```

```python
x = dataset[['country', 'year', 'status', 'adult_mortality', 'infant_deaths', 'alcohol', 'percentage_expenditure',
             'hepatitis_B', 'measles', 'bmi', 'under_5_deaths', 'polio', 'total_expenditure', 'diphtheria',
             'hiv_aids', 'gdp', 'population', 'thin_1-19_years', 'thin_5-9_years', 'income_comp_of_resources', 'schooling']] # independent variables/features are classified as x
y = dataset['life_expectancy'] # Target variable that we are trying to predict are classified as y
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(scaled_features, y, test_size = 0.3, random_state = 101)
```

#### LINEAR REGRESSION

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train, y_train)
```

```python
lr.intercept_
```

```python
coeff_df = pd.DataFrame(lr.coef_, x.columns, columns=['Coefficient'])
# create a dataframe showing the features and their corresponding coefficient
# lrm.coef is the data x.columns is the index

coeff_df
```

```python
pred = lr.predict(x_test)
plt.scatter(y_test,pred) #shows the correlation between pred and y_test

```

```python
sns.distplot((y_test - pred), bins=50); #shows d difference btwn actual values y_test and pred
```

```python
from sklearn import metrics
print('MAE: ', metrics.mean_absolute_error(y_test, pred))
print('MSE: ', metrics.mean_squared_error(y_test, pred))
print('RMSE: ', np.sqrt(metrics.mean_absolute_error(y_test, pred)))

print('explained variance score: ', metrics.explained_variance_score(y_test,pred)) # shows d rlatnship btwn actual values and pred

# the higher the explained_variance_score the more accurate the model.
```

```python
dataset.columns
```

```python
print('test score: ', lr.score(x_test, y_test))
print('train score: ', lr.score(x_train, y_train))
```
